## **Q1:-**
### **What is meant by time-dependent seasonal components?**

### **Ans:-**

### **Time-dependent seasonal components refer to patterns or fluctuations in data that occur regularly and predictably over time, typically within a single year or over multiple years. These components are characterized by their periodic nature, where the same patterns repeat at specific intervals. Time-dependent seasonal components are a crucial aspect of time series analysis, as they help analysts understand and model the underlying patterns in the data.**

## **Q6:-** 
### **What is a moving average (MA) model and how does it differ from other time series models?**

### **Ans:-**

### **The moving average model specifies that the output variable depends linearly on the current and various past values of a stochastic (imperfectly predictable) term. Rather than using the past values of the forecast variable in a regression, a moving average model uses past forecast errors in a regression-like model.**

## **Q7:-** 
### **What is a mixed ARMA model and how does it differ from an AR or MA model?**

### **Ans:-**

### **ARMA is the combination of the AR and MA models. ARMA models cover both aspects of AR and MA. The ARMA model predicts the future values based on both the previous values and errors. Thus ARMA has better performance than AR and MA models alone.**
